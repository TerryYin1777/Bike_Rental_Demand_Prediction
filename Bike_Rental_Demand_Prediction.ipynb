```python
#regression model 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
bike = pd.read_csv('bike_train.csv')
bike_f = bike[bike["hour"]==1]
y=bike_f['count']
x1 = bike_f.drop(['date','count','dayofweek','holiday','hour'],1)
x2 = sm.add_constant(x1) 
bike_model = sm.OLS(y,x2).fit() 
bike_model.summary()
```

```python
#regression test 2

bike_test = pd.read_csv('bike_test.csv')
bike_test_f = bike[bike["hour"]==1]
y_test = bike_test_f['count']
x1_test = bike_test_f.drop(['date','count','holiday','dayofweek','hour'],1)
x2_test = sm.add_constant(x1_test) 
betaHat=bike_model.params.values 
yhat = np.dot(x2_test,betaHat)
color = np.random.random(size=yhat.size)
b = y_test
plt1 = plt.plot(y_test,b,'r--')
plt.xlabel('actual value(count)')
plt.ylabel('predicted value(count)')
plt.title("2 o'clock for adjusted model")
plt.scatter(y_test,yhat)
plt.show()
```

```python
#regression test 2(as a whole)
#figure out 24 model parameters
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
bike = pd.read_csv('bike_train.csv')
betaHat = []
for i in range(0,23):
    bike_f = bike[bike["hour"]==i]
    y=bike_f['count']
    x1 = bike_f.drop(['date','count','dayofweek','holiday','hour'],1)
    x2 = sm.add_constant(x1) 
    bike_model = sm.OLS(y,x2).fit() 
    betaHat.append(bike_model.params.values)
bike2 = pd.read_csv('bike_test.csv')
for j in range(0,23):
    bike_test_f = bike[bike["hour"]==j]
    y_test = bike_test_f['count']
    x1_test = bike_test_f.drop(['date','count','holiday','dayofweek','hour'],1)
    x2_test = sm.add_constant(x1_test) 
    yhat = np.dot(x2_test,betaHat[j])
    color = np.random.random(size=yhat.size)
    plt.scatter(y_test,yhat)
b = y_test
plt1 = plt.plot(y_test,b,'r--')
plt.xlabel('actual value(count)')
plt.ylabel('predicted value(count)')
plt.title('model 2')
plt.show()
```

```python
#RMSE TESTING FOR MODEL 1
n = len(y_test)
RMSE2 = np.sqrt(sum((yhat-y_test)**2)/n)
print(RMSE2)
```
